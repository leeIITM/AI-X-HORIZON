getting the edge connected between the nodes of the graph

```python
import itertools
import numpy as np

classes = df["Pclass"].unique()
all_edges = np.array([], dtype=np.int32).reshape((0, 2))
for team in classes:
    # Use 'team' instead of 'classes' to filter the DataFrame
    class_df = df[df["Pclass"] == team]
    people = class_df["PassengerId"].values
    # Build all combinations, as all players are connected
    permutations = list(itertools.combinations(people, 2))
    edges_source = [e[0] for e in permutations]
    edges_target = [e[1] for e in permutations]
    team_edges = np.column_stack([edges_source, edges_target])
    all_edges = np.vstack([all_edges, team_edges])
# Convert to Pytorch Geometric format
edge_index = all_edges.transpose()
edge_index # [2, num_edges]
```
